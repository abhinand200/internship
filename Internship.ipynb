```python
from sklearn.preprocessing import LabelEncoder

# Create a copy of the DataFrame to avoid modifying the original data
df_encoded = df.copy()

# Initialize LabelEncoder
le = LabelEncoder()

# Iterate through columns and encode categorical features
for col in df_encoded.select_dtypes(include=['object']).columns:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()

# Create the heatmap
plt.figure(figsize=(16, 10))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap of All Columns')
plt.show()
```

**Feature Reduction**

```python
df=df.drop(['Gender','Warranty Extension','Preferred Visit Time','Membership Status'],axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the categorical columns
categorical_columns = ['Region','Product Category','Payment Method'] # Fixed typo here

# Instantiate the OneHotEncoder
encoder = OneHotEncoder()

# Perform one-hot encoding on the selected columns of the DataFrame
encoded_data = encoder.fit_transform(df[categorical_columns])

# Convert the encoded data to a DataFrame with appropriate column names
encoded_columns = encoder.get_feature_names_out(categorical_columns)
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoded_columns) # Convert to dense array

# Concatenate the encoded columns with the original dataset (dropping the original categorical columns)
data_encoded = pd.concat([df.drop(columns=categorical_columns), encoded_df], axis=1)

# Show the first few rows of the transformed dataset
data_encoded.head()
```

```python
x=data_encoded.drop('Loyalty Score',axis=1)
y=data_encoded['Loyalty Score']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

**Model Building**

SVR

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
y_pred_svr=svr.predict(x_test)
```

```python
from  sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
r2=r2_score(y_test,y_pred_svr)
mae=mean_absolute_error(y_test,y_pred_svr)
msq=mean_squared_error(y_test,y_pred_svr)
print('r2 score:',r2)
print('mae:',mae)
print('msq:',msq)
```

KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_knn)
msq=mean_squared_error(y_test,y_pred_knn)
r2=r2_score(y_test,y_pred_knn)
print('r2 score:',r2)
print('mae:',mae)
print('msq:',msq)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_dt)
msq=mean_squared_error(y_test,y_pred_dt)
r2=r2_score(y_test,y_pred_dt)
print('r2 score:',r2)
print('mae:',mae)
print('msq:',msq)
```

Random Fores

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_rf)
msq=mean_squared_error(y_test,y_pred_rf)
r2=r2_score(y_test,y_pred_rf)
print('r2 score:',r2)
print('mae:',mae)
print('msq:',msq)
```

LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred_lr=lr.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_lr)
msq=mean_squared_error(y_test,y_pred_lr)
r2=r2_score(y_test,y_pred_lr)
print('r2 score:',r2)
print('mae:',mae)
print('msq:',msq)
```

LinearRegression algorithm gives the best model
